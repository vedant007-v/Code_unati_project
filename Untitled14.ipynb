```python
from PIL import Image
import torch
import re
from transformers import Qwen2VLForConditionalGeneration, AutoProcessor

# Load model and processor
device = "cuda" if torch.cuda.is_available() else "cpu"
model = Qwen2VLForConditionalGeneration.from_pretrained(
    "Qwen/Qwen2-VL-2B-Instruct",
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
).to(device)
processor = AutoProcessor.from_pretrained("Qwen/Qwen2-VL-2B-Instruct")

ingredients_data = {
    "rice meal": {"safety": "Safe in moderate amounts", "alternative": "Brown rice, quinoa", "daily_limit": "Varies based on diet"},
    "edible vegetable oil": {"safety": "Limit intake due to fat content", "alternative": "Olive oil, coconut oil", "daily_limit": "Less than 30g"},
    "corn meal": {"safety": "Generally safe", "alternative": "Whole wheat flour, oat flour", "daily_limit": "Moderate consumption"},
    "chickpea flour": {"safety": "Good source of protein", "alternative": "Lentil flour, soy flour", "daily_limit": "Safe for regular consumption"},
    "iodized salt": {"safety": "Excess intake can raise blood pressure", "alternative": "Himalayan pink salt", "daily_limit": "Less than 5g"},
    "spices & condiments": {"safety": "Generally safe", "alternative": "Fresh herbs", "daily_limit": "Use as per taste"},
    "sugar": {"safety": "Excessive intake leads to obesity and diabetes", "alternative": "Honey, jaggery", "daily_limit": "25g (WHO recommendation)"},
    "acidity regulator": {"safety": "Some may cause digestive issues", "alternative": "Lemon juice, vinegar", "daily_limit": "Use sparingly"},
    "anticaking agent": {"safety": "Generally safe", "alternative": "Rice flour, silica gel", "daily_limit": "As per regulations"},
    "flavour enhancer": {"safety": "Can cause sensitivity in some people", "alternative": "Natural spices", "daily_limit": "Limited consumption"},
    "potatoes": {"safety": "Good source of carbohydrates", "alternative": "Sweet potatoes, yams", "daily_limit": "Safe in moderate amounts"},
    "vegetable oil": {"safety": "Limit due to fat content", "alternative": "Olive oil, avocado oil", "daily_limit": "Less than 30g"},
    "wheat flour": {"safety": "Contains gluten, avoid if intolerant", "alternative": "Almond flour, coconut flour", "daily_limit": "As per dietary needs"},
    "invert sugar": {"safety": "Similar to regular sugar, limit intake", "alternative": "Stevia, monk fruit", "daily_limit": "25g (WHO recommendation)"},
    "whey powder": {"safety": "Good source of protein", "alternative": "Plant-based protein", "daily_limit": "As per dietary needs"},
    "lactose": {"safety": "Avoid if lactose intolerant", "alternative": "Lactose-free dairy", "daily_limit": "Based on tolerance"},
    "leavening agents": {"safety": "Generally safe", "alternative": "Baking soda, yeast", "daily_limit": "Use in baking as needed"},
    "emulsifiers": {"safety": "Some may cause digestive issues", "alternative": "Natural emulsifiers like lecithin", "daily_limit": "Use sparingly"},
    "dough conditioner": {"safety": "Some contain sulfites, may cause sensitivity", "alternative": "Natural enzymes", "daily_limit": "Limited consumption"},
    "artificial flavors": {"safety": "Some may cause allergies", "alternative": "Natural extracts", "daily_limit": "Use sparingly"},
    "caramel color": {"safety": "Some types may contain sulfites", "alternative": "Cocoa powder, beet juice", "daily_limit": "Use in moderation"},
    "vitamin fortification": {"safety": "Generally beneficial", "alternative": "Whole foods", "daily_limit": "As per dietary needs"},
     "edible common salt": {"safety": "Safe in moderate amounts", "alternative": "Rock salt, Himalayan pink salt", "daily_limit": "Less than 5g (WHO recommendation)"},
    "cumin": {"safety": "Generally safe", "alternative": "Caraway seeds, coriander seeds", "daily_limit": "Moderate consumption"},
    "chilli": {"safety": "May cause irritation in excess", "alternative": "Paprika, black pepper", "daily_limit": "Moderate consumption"},
    "black pepper": {"safety": "Generally safe", "alternative": "White pepper, cayenne pepper", "daily_limit": "Moderate consumption"},
    "dry mango": {"safety": "Generally safe", "alternative": "Tamarind, lemon juice", "daily_limit": "Moderate consumption"},
    "dehydrated vegetable powder (onion, garlic)": {"safety": "Safe in moderate amounts", "alternative": "Fresh onion and garlic", "daily_limit": "Based on tolerance"},
    "natural and nature identical flavouring substances (garlic, black pepper)": {"safety": "Generally safe", "alternative": "Natural extracts of garlic and black pepper", "daily_limit": "Use sparingly"},
    "paprika extract (INS 160C)": {"safety": "Safe as a food coloring", "alternative": "Natural paprika", "daily_limit": "Follow food safety guidelines"}
}

# Helper functions
def normalize_ingredient(raw_ingredient):
    """Clean and standardize ingredient names"""
    # Remove percentages, codes, and special characters
    cleaned = re.sub(r"\(.*?\)|\d+%|[^a-zA-Z& ]", "", raw_ingredient)
    return cleaned.strip().lower()

def find_ingredient_match(ingredient, database):
    """Find best matching ingredient in database"""
    normalized = normalize_ingredient(ingredient)

    # Check exact matches first
    if normalized in database:
        return normalized

    # Check partial matches
    for db_ingredient in database:
        if db_ingredient in normalized or normalized in db_ingredient:
            return db_ingredient

        # Check word components
        db_words = db_ingredient.split()
        if all(word in normalized.split() for word in db_words):
            return db_ingredient

    return None

# Specify image path and user preference
image_path = "ingri.jpg"
user_preference = "low-sodium"

# Image processing
image = Image.open(image_path).convert("RGB").resize((448, 448))

# Create prompt
messages = [{
    "role": "user",
    "content": [
        {"image": image},
        {"text": f"Extract ONLY the ingredients list from this product label. List them with each item on a new line starting with '-'."}
    ]
}]

# Generate and process response
text_prompt = processor.apply_chat_template(messages, add_generation_prompt=True)
inputs = processor(text=[text_prompt], images=[image], return_tensors="pt").to(device)
output_ids = model.generate(**inputs, max_new_tokens=500)
full_response = processor.batch_decode(output_ids, skip_special_tokens=True)[0]

# Clean response
cleaned_response = full_response.split("assistant")[-1].strip() if "assistant" in full_response else full_response.strip()

# Parse ingredients
nutrition_keywords = {'calories', 'protein', 'carb', 'fat', 'sodium', 'vitamin', 'mineral', '%'}
detected_ingredients = [
    line.strip().split("-", 1)[1].strip()  # Remove leading dash
    for line in cleaned_response.split("\n")
    if line.strip().startswith("-") and
    not any(keyword in line.lower() for keyword in nutrition_keywords)
]

print("="*50)
print("Product Analysis Report")
print("="*50)
print("Extracted Ingredients:")
print("- " + "\n- ".join(detected_ingredients))

print("\n" + "="*50)
print("Ingredient Safety Analysis")
print("="*50)

# Analyze ingredients
for raw_ingredient in detected_ingredients:
    matched = find_ingredient_match(raw_ingredient, ingredients_data)

    if matched:
        info = ingredients_data[matched]
        print(f"\nIngredient: {raw_ingredient}")
        print(f"  Normalized Name: {matched.capitalize()}")
        print(f"  Safety: {info['safety']}")
        print(f"  Alternatives: {info['alternative']}")
        print(f"  Daily Limit: {info['daily_limit']}")
    else:
        print(f"\nIngredient: {raw_ingredient} - No safety data available")

print("\n" + "="*50)
print("Dietary Recommendation")
print("="*50)
print(f"For a {user_preference} diet:")
print("- Consider reducing sodium intake from salt and processed ingredients")
print("- Opt for fresh herbs instead of flavor enhancers")
print("- Monitor intake of oils and carbohydrates")
```
